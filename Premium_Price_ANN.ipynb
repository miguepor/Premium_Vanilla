```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from itertools import product
from sklearn.neural_network import MLPRegressor
```

Let us assume that we want to calculate the premium of a vanilla European style call option. It is well known that in this easy -though important- situation, and under some conditions (e.g. log normality of prices) a closed formula (Black Scholes) can be employed:

$$
C_{t}=S_{t} N\left(d_{1}\right)-X e^{-r(T-t)} N\left(d_{2}\right)
$$

Where:

$$
\begin{array}{c}
d_{1}=\frac{\log \left(\frac{S}{X}\right)+\left(r-0.5 \sigma^{2}\right)(T-t)}{\sigma \sqrt{T-t}} \\
d_{2}=d_{1}-\sigma \sqrt{T-t}
\end{array}
$$



```python
def Black_Scholes(S,X,T,r,sigma):
  """
    * S: Spot price 
    * X: Strike price 
    * T: Days to strike date divided by stock market days (252)
    * r: Risk free rate
    * sigma: Annual volatility of the returns of the stock 
  """

  d1= np.log(S/X) + (r-(0.5*(sigma**2))*(T))/sigma*np.sqrt(T)
  d2 = d1-sigma*(np.sqrt(T))
  value= S*norm.cdf(d1) - (X* np.e ** (-r*T) * norm.cdf(d2))
  return value
```

```python
Ss= np.arange(80,121,10)
Xs= np.arange(80,121,10)
sigma_s=np.arange(0.1,0.21,0.02)
rs=np.arange(0,0.11,0.02)
ts= np.arange(0.5,1.1,0.1)
```

Now we'll obtain the results of applying the Black-Scholes formula transforming the parameters we have just generated into a dataframe

```python
Cs_training = pd.DataFrame((product(Ss,Xs,ts,rs,sigma_s)),columns=("S","X","T","R","sigma"))
#We'll use the cartesian product of all the iterable elements in order to be able to make every single possible combination
Premium=Cs_training.apply(lambda a: Black_Scholes(a['S'],a['X'],a['T'],a['R'],a['sigma']),axis=1)
```

Now, we are going to compute the NN, we want to train. In instance, we are using a MLP. We are going calculate as well, the correlation coefficient between the output and the expected output

For weighting solver we'll use 'lbfgs' which for smaller datasets can converge faster and perform
better.

Activation function = relu

```python
X = Cs_training.values
Y = Premium.values

ann = MLPRegressor(hidden_layer_sizes=(10),solver="lbfgs",max_iter=500,batch_size=100)
ann.fit(X,Y)

print(f"The coefficient of determination R^2 of the prediction is = {round(ann.score(X,Y)*100,2)}%")
```

After having trained the NN, let's see how it performs with unseen values, comparing the output obtained in the ANN with the real premium value by BS formula. 





```python
S_test = 100
X_test = 100
T_test = .8
r_test = .05
sigma_test = .15 

premium_test= Black_Scholes(S_test,X_test,T_test,r_test,sigma_test)
premium_test_ann = ann.predict(np.c_[S_test,X_test,T_test,r_test,sigma_test]) #We need to convert the inputs in a Matrix
```

```python
print(f""" Test in formula= {premium_test}
      Test in ANN = {premium_test_ann}""")
```
